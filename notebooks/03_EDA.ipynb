```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme(style="whitegrid")

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/processed/cleaned_jobs.csv")

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe(include="all").T
```

```python
plt.figure(figsize=(12,6))

top_locations = df["location"].value_counts().head(15)

sns.barplot(
    x=top_locations.values,
    y=top_locations.index
)

plt.title("Top Hiring Cities")
plt.xlabel("Number of Jobs")
plt.ylabel("City")

plt.tight_layout()
plt.savefig("../outputs/figures/top_hiring_cities.png")
plt.show()
```

```python
plt.figure(figsize=(12,6))

top_companies = df["companyname"].value_counts().head(15)

sns.barplot(
    x=top_companies.values,
    y=top_companies.index
)

plt.title("Top Hiring Companies")
plt.xlabel("Number of Job Postings")
plt.ylabel("Company")

plt.tight_layout()

plt.savefig("../outputs/figures/top_companies.png")

plt.show()
```

```python
salary = df[df["maximumsalary"] > 0]["maximumsalary"]

plt.figure(figsize=(10,6))

sns.histplot(
    salary,
    bins=40,
    kde=True
)

plt.title("Maximum Salary Distribution")
plt.xlabel("Maximum Salary (INR)")
plt.ylabel("Number of Jobs")

plt.tight_layout()
plt.savefig("../outputs/figures/salary_distribution.png")

plt.show()
```

```python
salary = df[df["maximumsalary"] > 0]["maximumsalary"]

plt.figure(figsize=(10,2))

sns.boxplot(x=salary)

plt.title("Maximum Salary Boxplot")

plt.tight_layout()
plt.savefig("../outputs/figures/salary_boxplot.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df["avg_experience"], bins=20, kde=True)

plt.title("Average Experience Required")

plt.xlabel("Years")

plt.tight_layout()

plt.savefig("../outputs/figures/experience_distribution.png")

plt.show()
```

```python
from collections import Counter

skills = []

for row in df["tagsandskills"].dropna():

    for skill in str(row).split(","):

        skills.append(skill.strip())

counter = Counter(skills)

top_skills = pd.DataFrame(
    counter.most_common(20),
    columns=["Skill","Count"]
)

plt.figure(figsize=(12,7))

sns.barplot(
    data=top_skills,
    x="Count",
    y="Skill"
)

plt.title("Top Skills in Demand")

plt.tight_layout()

plt.savefig("../outputs/figures/top_skills.png")

plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.tight_layout()

plt.savefig("../outputs/figures/correlation_heatmap.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="avg_experience",
    y="maximumsalary"
)

plt.title("Maximum Salary vs Experience")

plt.tight_layout()

plt.savefig("../outputs/figures/salary_vs_experience.png")

plt.show()
```
