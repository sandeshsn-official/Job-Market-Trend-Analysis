```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/cleaned_jobs.csv")

print(df.shape)

df.head()
```

```python
df.columns.tolist()
```

```python
df["salary_band"] = pd.cut(
    df["maximumsalary"],
    bins=[0,300000,600000,1000000,2000000,10000000],
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High",
        "Premium"
    ]
)

df["salary_band"].value_counts()
```

```python
df["experience_category"] = pd.cut(
    df["avg_experience"],
    bins=[0,2,5,8,50],
    labels=[
        "Entry",
        "Mid",
        "Senior",
        "Expert"
    ]
)

df["experience_category"].value_counts()
```

```python
df["rating_category"] = pd.cut(
    df["aggregaterating"],
    bins=[0,3,4,5],
    labels=[
        "Average",
        "Good",
        "Excellent"
    ]
)

df["rating_category"].value_counts()
```

```python
df["salary_gap"] = (
    df["maximumsalary"] -
    df["minimumsalary"]
)

df[[
    "minimumsalary",
    "maximumsalary",
    "salary_gap"
]].head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_salary = LabelEncoder()

df["salary_band_encoded"] = le_salary.fit_transform(
    df["salary_band"].astype(str)
)

df.head()
```

```python
le_exp = LabelEncoder()

df["experience_encoded"] = le_exp.fit_transform(
    df["experience_category"].astype(str)
)

df.head()
```

```python
le_rating = LabelEncoder()

df["rating_encoded"] = le_rating.fit_transform(
    df["rating_category"].astype(str)
)

df.head()
```

```python
features = [
    "companyname",
    "title",
    "location",
    "minimumsalary",
    "maximumsalary",
    "salary_gap",
    "avg_experience",
    "aggregaterating",
    "reviewscount",
    "salary_band",
    "experience_category",
    "rating_category",
    "salary_band_encoded",
    "experience_encoded",
    "rating_encoded",
    "tagsandskills"
]

feature_df = df[features]

feature_df.head()
```

```python
feature_df.isnull().sum()
```

```python
feature_df["aggregaterating"] = feature_df["aggregaterating"].fillna(0)

feature_df["reviewscount"] = feature_df["reviewscount"].fillna(0)

feature_df["tagsandskills"] = feature_df["tagsandskills"].fillna("Unknown")

feature_df["salary_band"] = feature_df["salary_band"].fillna("Unknown")

feature_df["experience_category"] = feature_df["experience_category"].fillna("Unknown")

feature_df["rating_category"] = feature_df["rating_category"].fillna("Unknown")
```

```python
feature_df["aggregaterating"] = feature_df["aggregaterating"].fillna(0)
feature_df["reviewscount"] = feature_df["reviewscount"].fillna(0)
feature_df["tagsandskills"] = feature_df["tagsandskills"].fillna("Unknown")

# Add Unknown category before filling
feature_df["salary_band"] = (
    feature_df["salary_band"]
    .cat.add_categories(["Unknown"])
    .fillna("Unknown")
)

feature_df["experience_category"] = (
    feature_df["experience_category"]
    .cat.add_categories(["Unknown"])
    .fillna("Unknown")
)

feature_df["rating_category"] = (
    feature_df["rating_category"]
    .cat.add_categories(["Unknown"])
    .fillna("Unknown")
)
```

```python
feature_df.to_csv(
    "../data/processed/final_jobs.csv",
    index=False
)

print("Feature Engineering Completed!")

print(feature_df.shape)
```
