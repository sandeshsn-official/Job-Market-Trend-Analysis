```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/processed/master_jobs.csv")

print("Dataset Loaded Successfully!")
print(df.shape)
```

```python
df.head()
```

```python
df.info()
```

```python
missing = pd.DataFrame({
    "Missing Values": df.isnull().sum(),
    "Percentage": (df.isnull().sum()/len(df))*100
})

missing = missing[missing["Missing Values"]>0]

missing.sort_values("Missing Values", ascending=False)
```

```python
print("Duplicates Before:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:", df.duplicated().sum())
```

```python
threshold = len(df) * 0.60

df = df.dropna(axis=1, thresh=threshold)

print(df.shape)
```

```python
df.columns.tolist()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns

df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
categorical_cols = df.select_dtypes(include="object").columns

df[categorical_cols] = df[categorical_cols].fillna("Unknown")
```

```python
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)

df.columns.tolist()
```

```python
[col for col in df.columns if "salary" in col]
```

```python
text_columns = [
    "title",
    "company_name",
    "location"
]

for col in text_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
```

```python
# Create average experience

df["avg_experience"] = (
    df["minimumexperience"] +
    df["maximumexperience"]
) / 2

df[["minimumexperience",
    "maximumexperience",
    "avg_experience"]].head()
```

```python
df.to_csv("../data/processed/cleaned_jobs.csv", index=False)

print("Cleaned dataset saved successfully!")
```
